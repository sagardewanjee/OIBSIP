```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
#1.dataset loading
df=pd.read_csv("D:/OASIS DATA ANALYSIS/project2/ifood_df.csv")
```

```python
#2.Data Exploration
print(df.info()) #shows basic info
print(df.shape)  #shows numbers of rows and column
print(df.head(3)) #shows first 3 rows
print(df.tail(3)) #shows last 3 rows
```

```python
# Data Cleaning
print(df.isnull().sum()) #checking for any null or missing values
#since no null value is detected from any column so droping is not done

#checking for outliers using interquartile range 
for col in df.columns:
    # Calculate and mark outliers
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"\nOutliers in {col}:")
    print(f"Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")
    print(f"Number of outliers: {len(outliers)}")
    
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col} (Outlier Detection)')
    plt.xlabel(col)
    plt.show()

#transforming outliers using robust scaler method as I didn't want to remove outliers column
for col in df.columns:
    scaler=RobustScaler()
    df[f'ro_{col}']=scaler.fit_transform(df[[col]])

    plt.figure(figsize=(8,5))
    sns.boxplot(x=df[f'ro_{col}'])
    plt.title(f'Box Plot of {col} (outlier transformation)')
    plt.tight_layout()
    plt.show()
    
```

```python
#3.Data Description
df.to_csv("D:/OASIS DATA ANALYSIS/project2/ifood_df_transformed.csv")
print(df.describe())
```

```python
 #4.Customer Segmentation
#new features for customer segmentation
df['ro_Total_Spending'] = df[['ro_MntWines', 'ro_MntFruits', 'ro_MntMeatProducts', 
                          'ro_MntFishProducts', 'ro_MntSweetProducts', 'ro_MntGoldProds']].sum(axis=1)

df['ro_Total_Purchases'] = df[['ro_NumWebPurchases', 'ro_NumCatalogPurchases', 
                           'ro_NumStorePurchases', 'ro_NumDealsPurchases']].sum(axis=1)

cluster_features = ['ro_Income', 'ro_Total_Spending', 'ro_Total_Purchases', 
                   'ro_Recency', 'ro_NumWebVisitsMonth','ro_Age']
X_cluster = df[cluster_features]

# Elbow method to find optimal k
wcss = [] #within-cluster sum of square
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_cluster)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for optimal cluster')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Apply K-means clustering
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
df['Cluster'] = kmeans.fit_predict(X_cluster)

# Analyze clusters
cluster_analysis = df.groupby('Cluster')[cluster_features].mean()
print("\nCluster Analysis:\n", cluster_analysis)

#PCA for Visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_cluster)
df['PC1'] = pca_components[:, 0]
df['PC2'] = pca_components[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PC1', y='PC2', hue='Cluster', palette='Set2')
plt.title('Customer Segments (PCA Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()

# Visualize clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='ro_Total_Spending', y='ro_Income', hue='Cluster', palette='viridis')
plt.title('Customer Segments by Income and Spending (Scaled)')
plt.xlabel('Scaled Total Spending')
plt.ylabel('Scaled Income')
plt.show()
```

```python
#5.Insights and Recommendations

# Creating total spending column if not present
if 'MntTotal' not in df.columns:
    df['MntTotal'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 
                          'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Summarizing cluster characteristics
summary = df.groupby('Cluster')[['Income', 'MntTotal', 'Recency']].mean().reset_index()
summary = summary.sort_values('Income', ascending=False)


print("Cluster Insights and Recommendations:\n")

for index, row in summary.iterrows():
    cluster = int(row['Cluster'])
    income = row['Income']
    spending = row['MntTotal']
    recency = row['Recency']

    print(f"Cluster {cluster} Analysis:")
    print(f"  -> Avg Income     : {income:.2f}")
    print(f"  -> Avg Spending   : {spending:.2f}")
    print(f"  -> Avg Recency    : {recency:.2f} (lower means more recent activity)")

    # Insight and recommendation logic-based
    if income > 60000 and spending > 800 and recency < 30:
        print("Insight       : These customers are wealthy, spend a lot, and recently shopped. Theyre your most valuable group.")
        print("Recommendation: Offer them with special perks, loyalty rewards, and early access to products.\n")

    elif income < 30000 and spending < 300 and recency > 60:
        print("Insight       : This group hasn't bought in a while, and they dont spend much. Possibly price-conscious or inactive.")
        print("Recommendation: Try to bring them back with discounts, bundle deals, or reactivation emails.\n")

    elif income > 40000 and spending < 600:
        print("Insight       : They earn well but dont spend much yet. Theres room to grow their interest.")
        print("Recommendation: Suggest related or premium products. Promote what similar customers buy.\n")

    elif recency > 60:
        print("Insight       : These customers havent purchased in a long time.")
        print("Recommendation: Reach out with a 'No see longtime' message or limited-time offer to re-engage.\n")

    else:
        print("Insight       : These are your regular, average customers.")
        print("Recommendation: Keep them engaged with general offers and seasonal promotions.\n")
```
